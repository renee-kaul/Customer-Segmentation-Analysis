#### We are interested in only the products that are Chips

```python
##Creating a unique list of all the Product names within the data
unique = list(transdata.PROD_NAME.unique())
unique
```

```python
## finding the different variations (maybe due to typing erorrs) that exist of the word 'Chip'
## because we only care about the product 'Chips'.
## We do this because we need to create a unique list that only has chips product in it.
chips_var = list()
for x in unique:
    temp = x.strip().split()
    for ele in temp:
        if ele.startswith('ch') or ele.startswith('Ch'):
            chips_var = chips_var + [ele]
                
chips_var
```

```python
## Finding index of the unique elements in the list of products which match a criteria
## In this case, the criteria is that the word 'chip' and its variations occurs in the list.
def Unique_index(unique):
    count = -1
    chips= list()
    for x in unique:
        count +=1
        temp = x.strip().split()
        for ele in temp:
            #Inculcating all the variations of the word 'Chip' according to the list chips_var
            if ele.startswith('chp') or ele.startswith('chps') or ele.startswith('Chps')\
            or ele.startswith('Chip') or ele.startswith('chip') or ele.startswith('chips')\
            or ele.startswith('Chips') or ele.startswith('Chp'):
                chips = chips + [count]
    return chips

Chips = Unique_index(unique)
print(Chips, end = " ")
```

```python
## We create a unique list of the product chips
## The list has 37 elements 
chips_uni = list()
for ind in transdata.index:
    for elem in Chips:
        if ind == elem:
            chips_uni += [unique[ind]]
chips_uni
```

```python
## Creating new dataset that contains only values of PROD_NAME variable that are in chips_uni list    
new_transdata = pd.DataFrame(transdata[transdata.PROD_NAME.isin(chips_uni) == True])

## Dropping the column of DATE in the new transdata as we do not require it
new_trans = new_transdata.drop(['DATE'], axis = 1)
new_trans
```

### Initial Data Analysis:


```python
merge
#Creating a df for only mainstream-midage singles/couples
select = merge.loc[merge['PREMIUM_CUSTOMER']== 'Mainstream']
midage_main = select.loc[select['LIFESTAGE']=='MIDAGE SINGLES/COUPLES']
midage_main = midage_main[['AVG_SALES']]
#Creating a df for only Budget + Premium -midage singles/couples
select2 = merge.loc[merge['PREMIUM_CUSTOMER']== 'Premium']
select3 = merge.loc[merge['PREMIUM_CUSTOMER']== 'Budget']
select2 = pd.concat([select2,select3])
midage_pb = select2.loc[select2['LIFESTAGE']=='MIDAGE SINGLES/COUPLES']
midage_pb = midage_pb[['AVG_SALES']]
midage_pb2 = midage_pb.to_numpy()
midage_main2 = midage_main.to_numpy()
```

```python
#Drawing Q-Q plots for midage singles/couples - mainstream, premium + budget
import statsmodels.api as sm
import pylab
mm = list()
for x in midage_main2:
    mm += [float(x)]
mm = np.array(mm)
print(mm)
mpb = list()
for y in midage_pb2:
    mpb += [float(y)]
mpb = np.array(mpb)
print(mpb)
mm_plot = sm.qqplot(mm, line = '45')
mpb_plot = sm.qqplot(mpb, line = '45')

pylab.show()
```

We have used Q-Q plots because the sample size for both the datasets is large, i.e, n>50. We can notice in the above two graphs that the two datasets are not normally distributed from the Q-Q plots. Therefore, it violates the basic assumption of t-test that both datasets are normally distributed. If this assumption is violated, it isn't valid to use the p-values to draw conclusions about the differences between the samples. 

Therefore, it is important to normalise the data first. It can be done using a min-max technique. Here, each value within the dataset is subtracted by the minimum value of the dataset and then divided by the range. 

```python
'''Normalisation of the data using min-max appraoch, i.e, subtract it by minimum value and divide it 
by the range'''
midage_main_scaled = midage_main.copy()
midage_pb_scaled = midage_pb.copy()
# apply normalization techniques
for column in midage_main_scaled.columns:
    midage_main_scaled[column] = (midage_main_scaled[column] - midage_main_scaled[column].min()) / (midage_main_scaled[column].max() - midage_main_scaled[column].min())
print(midage_main_scaled)

for column in midage_pb_scaled.columns:
    midage_pb_scaled[column] = (midage_pb_scaled[column] - midage_pb_scaled[column].min()) / (midage_pb_scaled[column].max() - midage_pb_scaled[column].min())
print(midage_pb_scaled)
```

H0: The mean sales is equal between the two groups

HA: The mean sales is not equal between the two groups

The level of significance is 0.05.

```python
import scipy.stats as stats
'''finding the variance of both the datasets and taking a ratio. If it is <4, then we can keep equal 
variance as true which is by default in the ttest_ind'''
print(np.var(midage_main_scaled)/np.var(midage_pb_scaled))
stats.ttest_ind(a=midage_main_scaled, b=midage_pb_scaled)
```

Because the p-value of our test (0.12427918) is more than alpha = 0.05, we cannot reject the null hypothesis of the test. We do not have sufficient evidence to say that the mean sales of the two customer segments is different.

If this assumption is violated, it isn't valid to use the p-values to draw conclusions about the differences between the samples. 

Therefore, it is important to normalise the data first. It can be done using a min-max technique. Here, each value within the dataset is subtracted by the minimum value of the dataset and then divided by the range. 

```python
'''Normalisation of the data using min-max appraoch, i.e, subtract it by minimum value and divide it 
by the range'''
young_main_scaled = young_main.copy()
young_pb_scaled = young_pb.copy()
# apply normalization techniques
for column in young_main_scaled.columns:
    young_main_scaled[column] = (young_main_scaled[column] - young_main_scaled[column].min()) / (young_main_scaled[column].max() - young_main_scaled[column].min())
print(young_main_scaled)

for column in young_pb_scaled.columns:
    young_pb_scaled[column] = (young_pb_scaled[column] - young_pb_scaled[column].min()) / (young_pb_scaled[column].max() - young_pb_scaled[column].min())
print(young_pb_scaled)
```
